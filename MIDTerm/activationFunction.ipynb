```python
import matplotlib.pyplot as plt
import numpy as np
#identify activation function y=f(x)==>y=x
x=np.arange(-15,15)
y=x
plt.plot(x,y,c="green")
plt.title("identify")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#binary activation function y=f(x)==>y=x if x<=0
x=np.arange(-15,15)
y=[]
for i in range(x.shape[0]):
    if(x[i]>0):
        y.append(1)
    else:
        y.append(0)
plt.plot(x,y,c="green")
plt.title("Binary Step")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#Sigmoid Activation function y=1/(1+e^-x)
x=np.arange(-15,15)
y=[]
for i in range(x.shape[0]):
    y.append(1/(1+np.exp(-x[i])))
plt.plot(x,y,c="green")
plt.title("Sigmoid Activation Function")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def tanh(x):
    t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    return t
```

```python
x=np.arange(-15,15,0.01)
plt.ax=plt.subplots(figsize=(9,5))
plt.plot(x,tanh(x), color="green", linewidth=3, label="tanh")
plt.legend(loc="upper left", frameon=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def relu(x):
    y=max(0,x)
    return y
```

```python
x=np.arange(-15,15,0.01)
relu_output=[relu(i) for i in x]
plt.ax=plt.subplots(figsize=(9,5))
plt.plot(x,relu_output, color="green", linewidth=3, label="Relu")
plt.legend(loc="upper left", frameon=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

# for input 
x = np.linspace(-15, 15,100)

# output values 
y = elu(x)


plt.plot(x, y)
plt.title("ELU")
plt.xlabel("")
plt.ylabel("")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def selu(x, alpha=1.67326, scale=1.0507):
    
    return scale * np.where(x > 0, x, alpha * (np.exp(x) - 1))

# input 
x = np.linspace(-15, 15, 10)

# output 
y = selu(x)

# Plot the SELU activation function
plt.plot(x, y)
plt.title("SELU")
plt.xlabel("")
plt.ylabel("")
plt.grid()
plt.show()
```
